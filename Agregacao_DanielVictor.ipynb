Passo 2: Crie 3 Séries diferentes, cada um de tamanho 100 como a seguir:

1. The first a random number from 1 to 4
2. The second a random number from 1 to 3
3. The third a random number from 10,000 to 30,000

```python
s1 = pd.Series(np.random.randint(1, 4, size=100))
s2 = pd.Series(np.random.randint(1, 3, size=100))
s3 = pd.Series(np.random.randint(10000, 30000, size=100))
```

Passo 3: Vamos criar um DataFrame unindo Series por coluna.

```python
df = pd.DataFrame([s1, s2, s3]).T
df.head()
```

```python
df = df.rename(columns= {0:'bedrs', 1:'bathrs', 2:'price_sqr_meter'})
df.head()
```

Passo 5: Crie uma coluna DataFrame com os valores da Series 3 e atribua-o a 'bigcolumn'.
