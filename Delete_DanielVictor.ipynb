```python
import numpy as np
import pandas as pd
```

Passo 2: importe o dataset pelo link fornecido.

Passo 3: Atribua-o a uma variável chamada 'wine'.

```python
wine = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data")
```

Passo 4: Delete a primeira, quarta, sétima, nona, décima primeira, décima terceira e décima quarta colunas.

```python
wine = wine.drop(columns=wine.columns[[0, 3, 6, 8, 10, 12, 13]], axis=1)
```
