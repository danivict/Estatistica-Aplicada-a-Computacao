**2. Qual a diferença entre uma lista, um dicionário, um array numpy e uma series
pandas?**

1. Uma lista possui elementos e esses elementos possuem índices, diferente da tupla, listas são mutáveis.
2. Dicionários possuem chaves que podem ter dados atribuídos as mesmas, apresentados forma mais elegante que a lista.
3. Um array numpy é muito parecido com uma lista, porém, munipular os elementos se torna uma tarefa mais simples e admite elementos de apenas um tipo.
4. Com series pandas cada elemento recebe um índice em uma respectiva ordem.

**14. Conte o número de valores ausentes em cada coluna do df . Qual coluna tem o
número máximo de valores ausentes?**

```python
lista = []
colunas = []
mcoluna = ''
maior = 0
for i in df3:
    cont = 0
    colunas.append(i)
    for j in range(93):
        if str(df3[i][j]) == 'nan':
            cont+=1
    lista += str(cont)
for i in range(len(lista)):
    if int(lista[i]) > maior:
        maior = int(lista[i])
for i in range(len(colunas)):
    if int(lista[i]) == maior:
        mcoluna += colunas[i]+', '
c = 0
for i in df3:
    print(f'Coluna {i} apresenta {lista[c]} valores ausentes.')
    c += 1
print(f'\nAs colunas {mcoluna}têm o número máximo de valores ausentes')
```

**19. Obtenha as duas últimas linhas de df cuja soma da linha é maior que 100.**

```python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
```

```python
alt = df.tail().iloc[-1:5]
alt
conj = []              #duas ultimas linhas do dataframe cuja soma é maior que 100(cem)
for i in range(5): 
    conj.append(df.tail().iloc[-1-i:5-i].sum(axis=1))
for i in range(len(conj)):
    if conj[i][14-i]>100:
        print(f'A soma da linha {14-i} é maior que 100.')
```
