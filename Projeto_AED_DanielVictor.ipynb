Função para criação de classes.

````python
def dist_frequencia_classes(x, nc='sturges'):
    fa,classes= np.histogram(x, bins=nc) 
    fa = fa.astype(int)
    fr = np.around(fa/fa.sum()*100,decimals=2) 
    fc = np.cumsum(fa).astype(int) 
    classe_str = []
    for i in range(len(classes)):
        if i<=len(classes)-2:
            classe_str.append('{:2.2f} |-- {:2.2f}' .format(classes[i],classes[i+1]))
        else:
            break         
    return pd.DataFrame({'Classes':classe_str, 
                         'fi': fa,
                         'fr': fr,
                         'Fc': fc
                        }).set_index('Classes') 
````

Concatenação de todos os DataFrames gerados pela função.

````python
tabela = pd.concat([dist_frequencia_classes(vargas1.index, nc = 1), dist_frequencia_classes(jose.index, nc = 1),
           dist_frequencia_classes(eurico.index, nc = 1), dist_frequencia_classes(vargas2.index, nc = 1),
          dist_frequencia_classes(cafe.index, nc = 1), dist_frequencia_classes(nereu.index, nc = 1),
          dist_frequencia_classes(jk.index, nc = 1), dist_frequencia_classes(goulart.index, nc = 1),
          dist_frequencia_classes(castelo.index, nc = 1), dist_frequencia_classes(silva.index, nc = 1),
          dist_frequencia_classes(medici.index, nc = 1), dist_frequencia_classes(geisel.index, nc = 1),
          dist_frequencia_classes(figueiredo.index, nc = 1), dist_frequencia_classes(sarney.index, nc = 1),
          dist_frequencia_classes(collor.index, nc = 1), dist_frequencia_classes(itamar.index, nc = 1),
          dist_frequencia_classes(FHC.index, nc = 1), dist_frequencia_classes(lula.index, nc = 1),
          dist_frequencia_classes(dilma.index, nc = 1), dist_frequencia_classes(temer.index, nc = 1),
          dist_frequencia_classes(bolso.index, nc = 1)])
tabela = tabela.reset_index()
````
